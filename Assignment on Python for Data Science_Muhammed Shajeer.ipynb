# To Read the Dataset

```python
#to execute multiple codes in same program
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
#importing the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#to read the provided dataset from location
df = pd.read_excel(r"C:\Users\shajeer\Desktop\iris.xls")
```

# To Display the Columns in Dataset

```python
#to display the dataset
print("Dataset displayed Below");
#to diplay full dataset
pd.set_option('display.max_rows', None);
pd.set_option('display.max_columns', None);
#to display 
display(df);
```

# Calculate Mean of each column

```python
#To calculate the mean of columns,ie aong axis=1
val = df.mean(axis=0,numeric_only=True,skipna=True)
#values pulled and converted to a dataframe and then transposed for better viewing
pd.DataFrame(val,columns=['Mean']).T
```

# Null values Count and Details

```python
#variable defined to get the total count of null values
df2=df.isnull().sum()
#count of null values for each items (SL,SW,PL,PW) pulled and converted to a dataframe and then transposed for better viewing
pd.DataFrame(df2,columns=['Null Values']).T
#Displayed the total sum of count of null values
print("Total number of Null values : ",df2.sum(axis='index'))
```

# Graphical representation ( Bar Chart) - Total Sum for SL,SW,PL,PW

```python
#Executed sum in the data array to get sum of each column 
sum_df= df.sum()
#Removed the column with string values
sum_dfn=sum_df.drop(index=sum_df.index[4])
#Converted the sum of columns to dataframe and added "Total Sum" along axis 0
sum_summary=pd.DataFrame(sum_dfn,columns=['Total Sum'])
#data array for sum displayed
display(sum_summary)


#bar plot function called for plotting Total Sum along Y axis and items along X axis. Other graph variables defined
plt.figure(figsize=(6,6))
ax = sum_dfn.plot.bar(y='Total Sum', width=.8,rot=0, color='green')
plt.title('Total Sum',fontsize=12);
plt.xlabel("Item",fontsize=10,color="Black");
plt.ylabel("Total Sum",fontsize=10,color="Black");
plt.grid('true');
```

# Graphical representation ( Pie Chart) - Iris type

```python
#to count the total iris types in dataframe format
count=pd.DataFrame(df['Classification'].value_counts())
#To display the table
display(count)
#Plot pie to plot the pie chart
df['Classification'].value_counts().plot(kind='pie')
plt.title('Count of Iris types',fontsize=12);
plt.ylabel("",fontsize=10,color="Black");
```

# Graphical representation ( Histogram) - SL

```python
#Plot histogram to plot the histofgram of SL
plt.figure();
plt.figure(figsize=(6,6));
plt.hist(df['SL'],color="Green",rwidth=1,bins=25);
plt.title('Histogram of SL',fontsize=12);
plt.xlabel("SL",fontsize=12,color="Black");
plt.ylabel("Frequency",fontsize=12,color="Black");
plt.grid(True);
plt.show();

```
